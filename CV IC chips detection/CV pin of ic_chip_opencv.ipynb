```python
# put all the previous 3 steps in one function 
# gray scale conversion ==> applying gaussian blur ==> applying canny function

# OpenCV
# Pytorch



def Canny(image):
    # grey
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
    # gaussien blur
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    
    # canny detection
    canny = cv2.Canny(blur, 50, 150)
    
    return canny
```

```python
image = cv2.imread("C:/Users/ASUS/Desktop/ic_chips_images/images/frame2.png")
chip_image = np.copy(image)
canny = Canny(chip_image)

plt.imshow(canny)
plt.show()
```

```python
# Otsu's thresholding after Gaussian filtering
image = cv2.imread(("C:/Users/ASUS/Desktop/ic_chips_images/images/frame2.png"), 0)
blur = cv2.GaussianBlur(image,(5,5),0)
ret3,th3 = cv2.threshold(blur, 191, 191, cv2.THRESH_BINARY)

plt.imshow(th3)
plt.show()
```

```python
# final step: put it all together
# now combine the above immage containing just the lines in a black image with our original image

combo_image = cv2.addWeighted(image, 1, th3, 1, 1)

plt.imshow(combo_image)
plt.show()
```
